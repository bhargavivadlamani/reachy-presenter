```python
# Trussed API endpoint for GPT-3.5 Turbo chat completions
api_url = 'https://fauengtrussed.fau.edu/provider/generic/chat/completions'

# Your conversation history
conversation_history = [
    {'role': 'system', 'content': 'You are the worlds leading expert on ducks.'},
    {'role': 'user', 'content': 'What is the most interesting duck in the world and why?'}
]

# Set up the request headers
headers = {
    'Content-Type': 'application/json',
    'Authorization': f'Bearer {api_key}',
}

# Set up the request payload
data = {
    'model': 'gpt-4o',  
    'messages': conversation_history,
    'max_tokens': 100,
    'temperature': 0.5,
}

# Make the POST request to the OpenAI API
response = requests.post(api_url, json=data, headers=headers)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Print the generated text
    print(response.json()['choices'][0]['message']['content'])
else:
    # Print the error message
    print(f"Error: {response.status_code} - {response.text}")
```
